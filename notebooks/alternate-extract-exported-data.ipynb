```python
import zipfile
import tempfile
from pathlib import Path
```

```python
EXPORTED_ZIPFILE = Path("/Users/mjboothaus/icloud/Data/apple_health_export/export_23May2023.zip") 
```

```python
def extract_subdir(zip_ref, subdir, temp_dir):
    for member in zip_ref.namelist():
        if member.startswith(subdir):
            zip_ref.extract(member, temp_dir)
            # print(f"Extracted {member} to {temp_dir}")
    return None
```

```python
if EXPORTED_ZIPFILE.exists():
    with zipfile.ZipFile(EXPORTED_ZIPFILE, 'r') as zip_ref:
        with tempfile.TemporaryDirectory() as temp_dir:
            #zip_ref.extractall(temp_dir, members=['apple_health_export/workout-routes/*.gpx'])
            extract_subdir(zip_ref, 'apple_health_export/workout-routes/', temp_dir)
            print(f"Extracting {EXPORTED_ZIPFILE} to {temp_dir}")
            # Perform any operations on the extracted files here
            gpx_files = list((Path(temp_dir) / "apple_health_export" / "workout-routes").iterdir())
            for filename in gpx_files:
                print(filename)
            print(f"Extracted {len(gpx_files)} files")

# The temporary directory and its contents will be deleted when the block exits
```
